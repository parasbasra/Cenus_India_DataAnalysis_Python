# Data Analysis with visualization on Census India

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.show()
```

## importing the dataset

```python
data = pd.read_csv(r'C:\Users\Lenovo\Downloads\census_dataset.csv')
```

```python
data
```

```python
data.groupby('State_name')['Population'].sum().sort_values(ascending=False) #sum of population per state
```

```python
data.groupby('State_name')[['Hindus','Muslims','Christians','Sikhs','Buddhists','Jains']].sum().sort_values(by='Hindus',ascending=False)
#sum of population per state with religions , in > 1 columns while sorting use by=
```

## 1. Population Distribution
### Uttar Pradesh has the highest population followed by Maharashtra.

```python
state_population = data.groupby('State_name')['Population'].sum().sort_values(ascending=False)
state_population.plot(kind='bar', figsize=(12, 6), color='skyblue')
plt.title('Population Distribution by State')
plt.xlabel('State')
plt.ylabel('Population')
plt.show()
```

## 2. Gender Distribution in State
### Again Uttar Pradesh has highest Gender Distribution and Population.

```python
gender_distribution = data.groupby(['State_name'])[['Male', 'Female']].sum()
gender_distribution.plot(kind='bar', stacked=True, figsize=(14, 7))
plt.title('Gender Distribution by State')
plt.xlabel('State, District')
plt.ylabel('Population')
plt.show()
```

## 3. Literate Population
### Kerala is the most Literate state

```python
literacy_rate = (data['Literate'] / data['Population']) * 100
data['Literate_Rate'] = literacy_rate

literacy_rate_by_state = data.groupby('State_name')['Literate_Rate'].mean().sort_values(ascending=False)
literacy_rate_by_state.plot(kind='bar', figsize=(12, 6), color='lightgreen')
plt.title('Average Literacy Rate by State')
plt.xlabel('State')
plt.ylabel('Literate Rate (%)')
plt.show()
```

## 4. Workforce Analysis
### Himachal Pradesh has the highest Average Workforce.

```python
workforce_percentage = (data['Workers'] / data['Population']) * 100
data['Workforce_Percentage'] = workforce_percentage

workforce_by_state = data.groupby('State_name')['Workforce_Percentage'].mean().sort_values(ascending=False)
workforce_by_state.plot(kind='bar', figsize=(12, 6), color='orange')
plt.title('Average Workforce Percentage by State')
plt.xlabel('State')
plt.ylabel('Workforce Percentage (%)')
plt.show()
```

## 5. Religious Diversity
### Hindus being the maximum population and jains being the lowest.

```python
religious_distribution = data[['Hindus','Jains', 'Sikhs','Muslims', 'Buddhists', 'Christians']].sum()
religious_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=30, pctdistance=1, wedgeprops=dict(width=0.4), colors=['#FF9999', '#66B2FF', '#99FF99', '#FFCC99'])
plt.title('Religious Composition')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

## 6. Education Levels
### Most of the population has attained secondary education.

```python
education_distribution = data[['Secondary_Education', 'Higher_Education', 'Graduate_Education']].sum()
education_distribution.plot(kind='bar', color=['#FF9999', '#66B2FF', '#99FF99'])
plt.title('Total Education Level Distribution')
plt.xlabel('Education Level')
plt.xticks(rotation=0)
plt.ylabel('Population')
plt.show()
```

## 7. Age Groups

```python
age_groups_distribution = data[['Age_Group_0_29', 'Age_Group_30_49', 'Age_Group_50']]
sns.violinplot(data=age_groups_distribution, inner="quartile", palette="muted", split=True)
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Population')
plt.show()
```

## 8. Correlation Analysis

```python
correlation_matrix = data[['Literate', 'Workers', 'Secondary_Education', 'Higher_Education', 'Graduate_Education']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

## 9. Community wise Analysis

```python
community_distribution = data[['State_name', 'Hindus','Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains']].set_index('State_name')
community_distribution.plot(kind='area', stacked=True, figsize=(14, 7))
plt.title('Community-wise Population Distribution by State')
plt.xlabel('State')
plt.ylabel('Population')
plt.show()
```
